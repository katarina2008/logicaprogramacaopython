```python
#entrada
Ht = float(input("Informe quantas horas você trabalhou:"))
Mt = float(input("Informe quantos minutos você trabalhou: "))
GpH = float(input("Informe quanto você ganha por hora: "))

#processamento
H = (Mt/60)
HeM = (H+Ht)
OGpH = (HeM*GpH)
R = (OGpH*70/100)+OGpH

#saida
print("O valor que você receberà serà de: R$",R)
```

Gasto de gasolina



```python
#entrada
K = float(input("Informe quantos km irá percorrer: "))
CG = float(input("Informe quanto seU carro consome de gasolina: "))
PMG = float(input("Informe qual é o preço médio da gasolina: "))

#processamento
GM = (K/CG)*PMG
IV = (GM+GM)

#saida
print ("O custo da ida será de: $",GM)
print ("O cuso de ida e volta será de: $",IV)
```
